```python
"""
1) What is unsupervised learning in the context of machine learning?
Answer-
    Unsupervised learning is a type of machine learning where the model finds patterns and structures in data without using labeled output.
    Common tasks include clustering and dimensionality reduction.
"""
```

```python
"""
2) How does K-Means clustering algorithm work?
Answer-
    K-Means partitions data into K clusters by initializing centroids, assigning points to the nearest centroid, 
    recalculating centroids, and repeating this process until convergence.
"""
```

```python
"""
3) Explain the concept of a dendrogram in hierarchical clustering.
Answer-
    A dendrogram is a tree-like diagram that shows the hierarchical relationship between data points, 
    used to visualize how clusters are merged or split at each step.
"""
```

```python
"""
4) What is the main difference between K-Means and Hierarchical Clustering?
Answer-
    K-Means requires a predefined number of clusters and works by optimizing centroids, 
    while Hierarchical Clustering builds a tree of clusters without needing to predefine the number.
"""
```

```python
"""
5) What are the advantages of DBSCAN over K-Means?
Answer-
    DBSCAN can find clusters of arbitrary shapes, handles noise/outliers well, 
    and does not require specifying the number of clusters beforehand.
"""
```

```python
"""
6) When would you use Silhouette Score in clustering?
Answer-
    Silhouette Score is used to evaluate how well each point fits within its cluster; 
    it helps in determining the optimal number of clusters.
"""
```

```python
"""
7) What are the limitations of Hierarchical Clustering?
Answer-
    It is computationally expensive for large datasets and is not flexible once clusters are formed.
"""
```

```python
"""
8) Why is feature scaling important in clustering algorithms like K-Means?
Answer-
    K-Means relies on distance calculations, so unscaled features can dominate the results and skew clustering.
"""
```

```python
"""
9) How does DBSCAN identify noise points?
Answer-
    DBSCAN marks points as noise if they are not reachable by a minimum number of neighbors within a given distance (epsilon).
"""
```

```python
"""
10) Define inertia in the context of K-Means.
Answer-
    Inertia is the sum of squared distances between each point and its assigned cluster centroid; lower inertia indicates better clustering.
"""
```

```python
"""
11) What is the elbow method in K-Means clustering?
Answer-
    The elbow method plots inertia vs. number of clusters and identifies the "elbow point" where the rate of improvement drops, 
    suggesting an optimal K.
"""
```

```python
"""
12) Describe the concept of "density" in DBSCAN.
Answer-
    In DBSCAN, density refers to the number of points within a specified radius (epsilon); 
    dense areas form clusters, and sparse areas are considered noise.
"""
```

```python
"""
13) Can hierarchical clustering be used on categorical data?
Answer-
    Yes, but it requires using appropriate distance measures like Hamming distance or converting data into numerical format.
"""
```

```python
"""
14) What does a negative Silhouette Score indicate?
Answer-
    A negative Silhouette Score indicates that a sample may have been assigned to the wrong cluster.
"""
```

```python
"""
15) Explain the term "linkage criteria" in hierarchical clustering.
Answer-
    Linkage criteria define how the distance between clusters is calculated, e.g., single, complete, or average linkage.
"""
```

```python
"""
16) Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
Answer-
    K-Means assumes equal-sized spherical clusters, so it struggles with irregular shapes, densities, or imbalanced sizes.
"""
```

```python
"""
17) What are the core parameters in DBSCAN, and how do they influence clustering?
Answer-
    DBSCAN has two main parameters: epsilon (max distance to consider neighbors) and min_samples (minimum neighbors to form a dense region).
"""
```

```python
"""
18) How does K-Means++ improve upon standard K-Means initialization?
Answer-
    K-Means++ spreads out the initial centroids to improve clustering performance and avoid poor local minima.
"""
```

```python
"""
19) What is agglomerative clustering?
Answer-
    Agglomerative clustering is a type of hierarchical clustering that starts with individual points 
    and iteratively merges the closest pairs of clusters.
"""
```

```python
"""
20) What makes Silhouette Score a better metric than just inertia for model evaluation?
Answer-
    Silhouette Score evaluates both cohesion (within cluster) and separation (between clusters), 
    whereas inertia only measures compactness.
"""
```

```python
"""
21) Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot
Answer-
"""
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='x')
plt.title("K-Means Clustering on make_blobs Data")
plt.show()
```

```python
"""
29) Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
Answer-
"""
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=300, centers=4, random_state=42)
scores = []

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.bar(range(2, 6), scores)
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different k")
plt.show()
```

```python
"""
30) Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage
Answer-
"""
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

iris = load_iris()
Z = linkage(iris.data, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z, labels=iris.target)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
"""
31) Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries
Answer-
"""
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Decision boundary
h = 0.1
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("K-Means with Decision Boundaries")
plt.show()
```

```python
"""
34) Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot
Answer-
"""
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = StandardScaler().fit_transform(data.data)
inertias = []

for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("KMeans Inertia for K = 2 to 6")
plt.show()
```

```python
"""
36) Use the Wine dataset, apply DBSCAN after scaling the data, and count the number of clusters (excluding noise)
Answer-
"""
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

data = load_wine()
X = StandardScaler().fit_transform(data.data)
dbscan = DBSCAN(eps=1.5, min_samples=5)
labels = dbscan.fit_predict(X)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Number of clusters (excluding noise):", n_clusters)
```

```python
"""
40) Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D scatter plot
Answer-
"""
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

digits = load_digits()
X_pca = PCA(n_components=3).fit_transform(digits.data)
kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X_pca)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=labels, cmap='tab10')
ax.set_title("KMeans Clustering on Digits PCA (3D)")
plt.show()
```
