```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('spotify.csv')
df
```

```python
"""
Q2) What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram? 
Answer-
"""
plt.figure(figsize=(10,6))
sns.histplot(df['Popularity'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity Among Hip-hop Tracks', fontsize=16)
plt.xlabel('Popularity', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
```

```python
"""
Q3)  Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot?
Answer-
"""
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity', color='purple', alpha=0.6)
plt.title('Relationship Between Track Popularity and Duration', fontsize=16)
plt.xlabel('Duration (ms)', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.show()
```

```python
"""
Q4) Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot?
Answer-
"""
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='Artist', order=df['Artist'].value_counts().head(10).index, palette='viridis')

plt.title('Number of Tracks per Artist', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Number of Tracks', fontsize=14)
plt.show()
```

```python
"""
Q5) What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.
Answer-
"""
least_popular_tracks = df.sort_values(by='Popularity', ascending=True).head(5)

least_popular_tracks[['Artist', 'Track Name', 'Popularity']]
```

```python
"""
Q6) Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average 
    popularity for each artist.
Answer-
"""

top_5_artists = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False).head(5).index
top_5_artists
```

```python

df_top_5_artists = df[df['Artist'].isin(top_5_artists)]

average_popularity = df_top_5_artists.groupby('Artist')['Popularity'].mean()

average_popularity.sort_values(ascending=False)
```

```python
"""
Q7) For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.
Answer-
"""
top_5_artists = df.groupby('Artist')['Popularity'].sum().nlargest(5).index

df_top_5_artists = df[df['Artist'].isin(top_5_artists)]

most_popular_tracks = df_top_5_artists.loc[df_top_5_artists.groupby('Artist')['Popularity'].idxmax()]

most_popular_tracks[['Artist', 'Track Name', 'Popularity']]
```

```python
"""
Q8) Visualize relationships between multiple numerical variables simultaneously using a pair plot.
Answer-
"""
numerical_columns = ['Popularity', 'Duration (ms)']

sns.pairplot(df[numerical_columns])

plt.suptitle('Pair Plot of Popularity and Duration (ms)', y=1.02, fontsize=14)
plt.show()
```

```python
"""
Q9)  Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.
Answer-
"""
top_artists = df['Artist'].value_counts().nlargest(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

plt.figure(figsize=(12, 8))
sns.boxplot(data=df_top_artists, x='Artist', y='Duration (ms)', palette='viridis')

plt.title('Variation in Track Duration for Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Track Duration (ms)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_artists, x='Artist', y='Duration (ms)', palette='viridis')

plt.title('Variation in Track Duration for Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Track Duration (ms)', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
"""
Q10) How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.
Answer-
"""
top_artists = df['Artist'].value_counts().nlargest(10).index
df_top_artists = df[df['Artist'].isin(top_artists)]

plt.figure(figsize=(12, 8))
sns.swarmplot(data=df_top_artists, x='Artist', y='Popularity', palette='viridis')

plt.title('Distribution of Track Popularity for Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Track Popularity', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top_artists, x='Artist', y='Popularity', palette='viridis')

plt.title('Distribution of Track Popularity for Top 10 Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Track Popularity', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
