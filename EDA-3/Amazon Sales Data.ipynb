```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('amazon.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
"""
Q1) What is the average rating for each product category?
Answer-
"""
df.columns
```

```python
df.rating.unique()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.rating.unique()
```

```python
df['rating'] = df['rating'].astype(float)
df['rating'].dtype
```

```python
average_rating_by_category = df.groupby('category')['rating'].mean().reset_index()
average_rating_by_category
```

```python
"""
Q2) What are the top rating_count products by category?
Answer-
"""
top_rating_count_by_category = df.sort_values(by=['category', 'rating_count'], ascending=[True, False])
top_products_by_category = top_rating_count_by_category.groupby('category').head(1)
top_products_by_category[['category', 'product_name', 'rating_count']]
```

```python
"""
Q3) What is the distribution of discounted prices vs. actual prices?
Answer-
"""
df.dtypes
```

```python
df.actual_price.unique()
```

```python
df.actual_price = df.actual_price.str.replace('₹' , '').str.replace(',' , '')
```

```python
df.actual_price = df.actual_price.astype('float')
```

```python
df.discounted_price = df.discounted_price.str.replace('₹' , '').str.replace(',' , '')
```

```python
df.discounted_price = df.discounted_price.astype('float')
```

```python
df.dtypes
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['actual_price'], label='Actual Price', fill=True, color='blue', alpha=0.5)
sns.kdeplot(df['discounted_price'], label='Discounted Price', fill=True, color='green', alpha=0.5)

plt.xlabel('Price')
plt.ylabel('Density')
plt.title('KDE Plot: Discounted vs Actual Prices')
plt.legend()
plt.show()
```

```python
"""
Q4) How does the average discount percentage vary across categories?
Answer-
"""
df.dtypes
```

```python
df.discount_percentage.unique()
```

```python
df.discount_percentage = df.discount_percentage.str.replace('%', '')
```

```python
df.discount_percentage = df.discount_percentage.astype('int')
```

```python
rename_dict = {
    'Computers&Accessories|Accessories&Peripherals|Cables&Accessories|CableConnectionProtectors': 'Cable Protectors',
    'Electronics|Headphones,Earbuds&Accessories|Earpads': 'Earpads',
    'Electronics|Mobiles&Accessories|MobileAccessories|Décor|PhoneCharms': 'Phone Charms',
    'Electronics|Headphones,Earbuds&Accessories|Adapters': 'Adapters',
    'Computers&Accessories|Accessories&Peripherals|Keyboards,Mice&InputDevices|Keyboard&MiceAccessories|DustCovers': 'Dust Covers',
    'Electronics|Mobiles&Accessories|MobileAccessories|Mounts|Shower&WallMounts': 'Wall Mounts',
    'Computers&Accessories|Components|InternalHardDrives': 'Hard Drives',
    'Computers&Accessories|Accessories&Peripherals|Adapters|USBtoUSBAdapters': 'USB Adapters',
    'Electronics|Mobiles&Accessories|MobileAccessories|Stands': 'Stands',
    'Electronics|Headphones,Earbuds&Accessories|Cases': 'Cases'
}

# Replace long category names with short names
df_filter = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).head(10)
df_filter = df_filter.rename(index=rename_dict)
print(df_filter)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot( x=df_filter.index, y= df_filter.values, palette='deep')
plt.xticks(rotation = 45)
plt.title('Top 10 Categories with Highest Average Discounts', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Discount Percentage (%)', fontsize=12)
plt.show()
```

```python
"""
Q5) What are the most popular product names?
Answer-
"""
# Shorten to the first 4 words
df['short_product_name'] = df['product_name'].apply(lambda x: ' '.join(x.split()[:4]))
popular_product = df.short_product_name.value_counts().head(10)
popular_product
```

```python
import matplotlib.pyplot as plt

# Bar plot for popular products
plt.figure(figsize=(10, 6))
popular_product.plot(kind='bar', color='lightgreen', edgecolor='black')

# Add labels and title
plt.title('Top 10 Most Popular Products', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Rotate labels and adjust layout
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
"""
Q6) What are the most popular product keywords?
Answer-
"""
from collections import Counter
import re

# Combine all product names into one text
all_text = ' '.join(df['product_name'])

# Extract keywords (filter only alphabetic words, ignoring case)
keywords = re.findall(r'\b[a-zA-Z]{3,}\b', all_text)

keyword_counts = Counter(keywords)

keyword_df = pd.DataFrame(keyword_counts.items(), columns=['Keyword', 'Count']).sort_values(by='Count', ascending=False)

print(keyword_df.head(10))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Keyword', data=keyword_df.head(10), palette='viridis')
plt.title('Top 10 Keywords in Product Names')
plt.xlabel('Count')
plt.ylabel('Keyword')
plt.show()
```

```python
"""
Q7)  What are the most popular product reviews?
Answer-
"""
review_counts = df['review_title'].value_counts().head(10)
review_counts
```

```python
"""
Q8)  What is the correlation between discounted_price and rating?
Answer-
"""
df[['discounted_price', 'rating']].isnull().sum()
```

```python
df.dropna(subset=['rating'], inplace=True)
```

```python
correlation = df[['discounted_price', 'rating']].corr()
correlation 
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap: Discounted Price vs Rating')
plt.show()
```

```python

"""
Q9)  What are the Top 5 categories based on the highest ratings?
Answer-
"""
category_ratings = df.groupby('category')['rating'].mean()

top_categories = category_ratings.sort_values(ascending=False).head(5)

top_categories
```

```python
rename_dict = {'Computers&Accessories|NetworkingDevices|NetworkAdapters|PowerLANAdapters' : 'Computers&Accessories|NetworkingDevices',
                'Electronics|Cameras&Photography|Accessories|Film' : 'Electronics|Cameras&Photography',
                'Computers&Accessories|Components|Memory ' : 'Computers&Accessories|Components',
                'Electronics|HomeAudio|MediaStreamingDevices|StreamingClients' : 'Electronics|HomeAudio'}


top_categories = top_categories.rename(index=rename_dict)
top_categories
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')

# Set plot labels and title
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Top 5 Categories Based on Average Ratings')
plt.xticks(rotation=45, ha='right')  # Rotate the category names for better visibility
plt.show()
```
