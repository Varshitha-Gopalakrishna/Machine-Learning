```python
"""
Q1) What does R-squared represent in a regression model?
Answer-
    R-squared (or the coefficient of determination) measures how well the regression model explains the variation in the dependent variable (y).
    What It Represents:
    Explained Variance: It tells you how much of the total variability in y is explained by the independent variables (x).
    Scale: R-squared ranges from 0 to 1:
           R2 =1: Perfect fit (model explains 100% of the variance).
           R2 =0: Model explains none of the variance.

"""
```

```python
"""
Q2) What are the assumptions of linear regression?
Answer-
    Assumptions of Linear Regression:
    Linearity:

    The relationship between the independent variable(s) and the dependent variable is linear.
    Example: A straight-line relationship, y=mx+c.
    Independence:

    Observations are independent of each other.
    There’s no correlation between errors (e.g., no patterns in residuals over time).
    Homoscedasticity:

    The variance of residuals (errors) is constant across all levels of the independent variable(s).
    Example: Error spread should not increase or decrease as x changes.
    Normality of Residuals:

    Residuals (differences between predicted and actual values) should follow a normal distribution.
    No Multicollinearity:

    Independent variables should not be highly correlated with each other.
    High correlation can distort the importance of variables.
"""
```

```python
"""
Q3) What is the difference between R-squared and Adjusted R-squared?
Answer-
    R-squared - 
    1. Measures how well the model explains the variation in the dependent variable.
    2. Always increases or stays the same, even if the new predictors do not improve the model.
    3. Does not penalize for adding more predictors.
    4. Good for evaluating the overall fit of the model.
    5. Value Range is Between 0 and 1.

    Adjusted R-squared - 
    1. Adjusted version of r squared that accounts for the number of predictors in the model.
    2. Only increases if the new predictors significantly improve the model.
    3. Penalizes for adding irrelevant predictors by adjusting for degrees of freedom.
    4. Better for comparing models with a different number of predictors.
    5. Value Range Can decrease and may be less than r squared
"""
```

```python
"""
Q4) Why do we use Mean Squared Error (MSE)?
Answer-
    Mean Squared Error (MSE) is commonly used in regression tasks to measure the accuracy of a model by quantifying the difference 
    between predicted and actual values.
    If MSE is small, the predictions are close to the actual values.
    If MSE is large, the model is making large errors and needs improvement.

"""
```

```python
"""
Q5)  What does an Adjusted R-squared value of 0.85 indicate?
Answer-
    An Adjusted R-squared value of 0.85 means that 85% of the variation in the dependent variable (target) is explained by the 
    independent variables (predictors) in the model, after accounting for the number of predictors and their relevance.
    The remaining 15% of the variation is due to factors not captured by the model (e.g., noise or other variables not included).
    An Adjusted R-squared of 0.85 indicates a strong, reliable model that is likely well-specified without overfitting. 

"""
```

```python
"""
Q6) How do we check for normality of residuals in linear regression?
Answer-
    The normality of residuals is a key assumption in linear regression.
    1. Plot a histogram of residuals and look for a bell-shaped curve.
    2. Check skewness (asymmetry) and kurtosis (peakedness). Both should be close to 0 for normal residuals.
    If residuals deviate significantly from normality, consider transformations or other models.
    
"""
```

```python
"""
Q7) What is multicollinearity, and how does it impact regression?
Answer-  
    Multicollinearity occurs when two or more independent variables in a regression model are highly correlated.  
    1. It makes it difficult to determine the individual effect of each variable on the dependent variable.  
    2. Coefficients of the correlated variables may become unstable and sensitive to small changes in data.  
    3. It inflates the standard errors, reducing the statistical significance of predictors.  
    To address multicollinearity, consider removing highly correlated variables or using techniques like Principal Component Analysis 
    (PCA).

"""
```

```python
"""
Q8) What is Mean Absolute Error (MAE)?  
Answer-  
    Mean Absolute Error (MAE) is a metric used to measure the average magnitude of errors in a set of predictions.  
    1. It is calculated as the mean of the absolute differences between predicted and actual values.  
    2. Formula: MAE = (1/n) * Σ |Actual - Predicted|.  
    3. MAE is easy to interpret, with lower values indicating better model performance.  
    It does not penalize large errors as heavily as Mean Squared Error (MSE).

"""
```

```python
"""
Q12) What does a high R-squared value mean?  
Answer-  
    A high R-squared value indicates that a significant proportion of the variance in the dependent variable is explained by the independent variables in the model.  
    1. **Interpretation**: It shows the goodness of fit, i.e., how well the model explains the observed data.  
    2. **Value Range**: R-squared ranges from 0 to 1, where values closer to 1 indicate better model performance.  
    3. **Limitations**:  
       - A high R-squared does not guarantee that the model is accurate or free from biases.  
       - It can be artificially inflated by adding irrelevant variables (overfitting).  
    Always consider other metrics, like adjusted R-squared, for a comprehensive evaluation.

"""
```

```python
"""
Q13) What happens if linear regression assumptions are violated?  
Answer-  
    Violating linear regression assumptions can lead to unreliable and biased results. The impact depends on the specific assumption violated:  
    1. **Linearity**: If the relationship is not linear, predictions may be inaccurate.  
    2. **Independence**: Violations can lead to autocorrelation, affecting hypothesis testing and confidence intervals.  
    3. **Homoscedasticity**: If residuals have unequal variance, standard errors may be biased, leading to incorrect significance tests.  
    4. **Normality of Residuals**: Violations affect the reliability of confidence intervals and hypothesis tests.  
    5. **No Multicollinearity**: High multicollinearity inflates standard errors and makes coefficient estimates unstable.  
    Addressing these violations may involve transformations, alternative models, or robust techniques.

"""
```

```python
"""
Q14) How can we address multicollinearity in regression?  
Answer-  
    Multicollinearity can be addressed using the following methods:  
    1. **Remove Highly Correlated Variables**: Drop one of the correlated variables to simplify the model.  
    2. **Combine Variables**: Use techniques like Principal Component Analysis (PCA) to create uncorrelated components.  
    3. **Regularization**: Apply techniques like Ridge Regression or Lasso Regression to reduce the impact of multicollinearity by penalizing large coefficients.  
    4. **Variance Inflation Factor (VIF)**: Identify variables with high VIF values (usually > 10) and remove or transform them.  
    5. **Data Collection**: Collect more data to reduce the correlation between variables.  
    These approaches improve model stability and interpretability.

"""
```

```python
"""
Q15) How can feature selection improve model performance in regression analysis?  
Answer-  
    Feature selection helps improve model performance by selecting the most relevant variables, reducing noise, and simplifying the model.  
    1. **Reduces Overfitting**: By removing irrelevant or redundant features, feature selection decreases the chance of overfitting and increases model generalization.  
    2. **Improves Accuracy**: Selecting the most important features allows the model to focus on the relevant data, potentially leading to better predictions.  
    3. **Enhances Interpretability**: A simpler model with fewer features is easier to understand and explain.  
    4. **Reduces Computational Cost**: Fewer features result in faster training and prediction times, making the model more efficient.  
    Common methods of feature selection include filter methods, wrapper methods, and embedded methods like Lasso.

"""
```

```python
"""
Q16) How is Adjusted R-squared calculated?  
Answer-  
    Adjusted R-squared is a modified version of R-squared that adjusts for the number of predictors in the model. It is calculated using the following formula:  
    1. **Formula**:  
       Adjusted R-squared = 1 - [(1 - R²) * (n - 1) / (n - p - 1)]  
       Where:  
       - **R²** is the R-squared value,  
       - **n** is the number of data points,  
       - **p** is the number of predictors.  
    2. **Interpretation**:  
       - Adjusted R-squared accounts for the number of predictors and helps prevent overfitting by penalizing the addition of irrelevant variables.  
       - It provides a more accurate measure of the goodness of fit when comparing models with different numbers of predictors.

"""
```

```python
"""
Q17) Why is MSE sensitive to outliers?  
Answer-  
    Mean Squared Error (MSE) is sensitive to outliers because:  
    1. **Squaring the Errors**: MSE calculates the average of the squared differences between actual and predicted values. Squaring magnifies large errors, making the model more sensitive to extreme values.  
    2. **Weight of Large Errors**: A small number of large errors (outliers) can significantly increase the MSE, even if the majority of predictions are accurate.  
    3. **Impact on Model Fit**: The presence of outliers can distort the model's fit, leading to a higher error value and potentially skewing model performance metrics.  
    To reduce the impact of outliers, alternative metrics like Mean Absolute Error (MAE) or robust regression techniques can be used.

"""
```

```python
"""
Q18) What is the role of homoscedasticity in linear regression?  
Answer-  
    Homoscedasticity refers to the assumption that the variance of the residuals (errors) is constant across all levels of the independent variable(s).  
    1. **Constant Variance**: In linear regression, homoscedasticity ensures that the residuals are evenly spread across the range of fitted values, implying no systematic pattern in the errors.  
    2. **Importance**:  
       - If residuals have constant variance, the model's predictions are reliable, and the statistical tests (e.g., t-tests, F-tests) are valid.  
       - Homoscedasticity indicates that the model's prediction errors are not influenced by the magnitude of the predicted values.  
    3. **Impact of Violation**: If homoscedasticity is violated (i.e., heteroscedasticity), it can lead to biased parameter estimates, incorrect standard errors, and invalid statistical tests.  
    Methods like weighted least squares or transforming variables can be used to address heteroscedasticity.

"""
```

```python
"""
Q19) What is Root Mean Squared Error (RMSE)?  
Answer-  
    Root Mean Squared Error (RMSE) is a metric used to measure the difference between the predicted and actual values in a regression model.  
    1. **Definition**: It is the square root of the average of the squared differences between predicted and actual values.  
    2. **Formula**:  
       RMSE = √[(1/n) * Σ (Actual - Predicted)²]  
       Where:  
       - **n** is the number of data points,  
       - **Actual** is the observed value,  
       - **Predicted** is the model's predicted value.  
    3. **Interpretation**:  
       - RMSE gives an estimate of the magnitude of the error and is in the same units as the target variable, making it easier to interpret.  
       - Lower RMSE values indicate better model performance.  
    RMSE penalizes large errors more than smaller ones due to the squaring of the differences.

"""
```

```python
"""
Q20) Why is pickling considered risky?  
Answer-  
    Pickling can be considered risky due to the following reasons:  
    1. **Security Risks**: Unpickling data from untrusted sources can lead to arbitrary code execution, making it vulnerable to attacks (e.g., remote code execution).  
    2. **Compatibility Issues**: Pickled objects may not be compatible across different Python versions or environments, leading to issues when loading objects saved in different versions.  
    3. **Corruption of Data**: If the pickled file is corrupted, it may lead to errors or loss of data when attempting to unpickle.  
    4. **Limited Portability**: Pickled objects are Python-specific and cannot be easily shared with other programming languages.  
    5. **Unpredictable Behavior**: Pickling complex or custom objects with external dependencies may lead to unexpected behavior upon unpickling, especially if class definitions change over time.  
    To mitigate risks, it is advised to use safer serialization formats like JSON or to limit unpickling to trusted sources.

"""
```

```python
"""
Q22) What is heteroscedasticity, and why is it a problem?  
Answer-  
    Heteroscedasticity occurs when the variance of the residuals (errors) in a regression model is not constant across all levels of the independent variable(s).  
    1. **Problem**:  
       - It violates one of the key assumptions of linear regression, which assumes homoscedasticity (constant variance of errors).  
       - Heteroscedasticity can lead to biased coefficient estimates, incorrect standard errors, and unreliable statistical tests.  
    2. **Impact**:  
       - Affects the significance of the model's predictors.  
       - Can lead to misleading conclusions about the relationship between the variables.  
    To address heteroscedasticity, you can transform the data, use robust standard errors, or apply weighted least squares regression.

"""
```

```python
# PRACTICAL QUESTIONS

"""
Q1) Write a Python script to visualize the distribution of errors (residuals) for a multiple linear regression model using Seaborn's "diamonds" dataset.
Answer-
"""
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

data = sns.load_dataset('diamonds')

data = data[['price', 'carat', 'depth', 'table', 'x', 'y', 'z']].dropna()

X = data[['carat', 'depth', 'table', 'x', 'y', 'z']]
Y = data['price']

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

residuals = model.resid

# Plot the distribution of residuals using Seaborn
sns.histplot(residuals, kde=True, color='blue', bins=30)

plt.title('Distribution of Residuals for Multiple Linear Regression')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
"""
Q2) Write a Python script to calculate and print Mean Squared Error (MSE), Mean Absolute Error (MAE), and Root Mean Squared Error 
     (RMSE) for a linear regression model.
Answer-
"""

import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing()
X = california_housing.data
Y = california_housing.target

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

# Calculate MSE, MAE, and RMSE
mse = mean_squared_error(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mse)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
"""
Q33) Write a Python script to check if the assumptions of linear regression are met. Use a scatter plot to check linearity, residuals
     plot for homoscedasticity, and correlation matrix for multicollinearity.
Answer-
"""
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the California housing dataset
from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing()
X = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
Y = california_housing.target

# 1. Scatter plot to check linearity (plotting one feature vs target)
sns.scatterplot(x=X['AveRooms'], y=Y)
plt.title('Scatter Plot to Check Linearity')
plt.xlabel('Average Rooms')
plt.ylabel('Target (House Value)')
plt.show()

# 2. Fit a linear regression model to check residuals (homoscedasticity)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, Y_train)

# Residuals
residuals = Y_test - model.predict(X_test)

# Plot residuals to check for homoscedasticity
sns.residplot(x=model.predict(X_test), y=residuals, lowess=True, color='blue')
plt.title('Residuals Plot to Check Homoscedasticity')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# 3. Correlation matrix to check multicollinearity
corr_matrix = X.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix to Check Multicollinearity')
plt.show()
```

```python
"""
Q4) Write a Python script that creates a machine learning pipeline with feature scaling and evaluates the performance of different regression models.
Answer-
"""
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# Load dataset (using California housing dataset as an example)
from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing()
X = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
Y = california_housing.target

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Define models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso()
}

# Create pipeline with feature scaling
for name, model in models.items():
    pipeline = make_pipeline(StandardScaler(), model)
    pipeline.fit(X_train, Y_train)
    Y_pred = pipeline.predict(X_test)
    
    # Calculate MSE
    mse = mean_squared_error(Y_test, Y_pred)
    print(f'{name} MSE: {mse}')

```

```python
"""
Q9) Write a Python script that fits a polynomial regression model (degree 2) to a dataset.
Answer-
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Create synthetic dataset (one feature)
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Feature: random values between 0 and 10
Y = 2 * X**2 + 3 * X + 5 + np.random.randn(100, 1) * 5  # Quadratic relation with noise

# Transform the features for polynomial regression (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Create and fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict new values
Y_pred = model.predict(X_poly)

# Plot the data points and polynomial regression curve
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Polynomial regression line')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()
```

```python
"""
Q11) Write a Python script that fits polynomial regression models of different degrees to a synthetic dataset and compares their performance.
Answer-
"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Feature: random values between 0 and 10
Y = 2 * X**2 + 3 * X + 5 + np.random.randn(100, 1) * 5  # Quadratic relation with noise

# Function to fit and plot polynomial regression for different degrees
def polynomial_regression(degree):
    poly = PolynomialFeatures(degree)
    X_poly = poly.fit_transform(X)
    model = LinearRegression()
    model.fit(X_poly, Y)
    
    # Predict and calculate MSE
    Y_pred = model.predict(X_poly)
    mse = mean_squared_error(Y, Y_pred)
    
    # Plot the data points and polynomial regression curve
    plt.plot(X, Y_pred, label=f'Degree {degree}')
    return mse

# Plotting the data points and comparing polynomial regression of different degrees
plt.scatter(X, Y, color='blue', label='Data points')
mse_values = []
for degree in [1, 2, 3, 4]:
    mse = polynomial_regression(degree)
    mse_values.append(mse)

plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Polynomial Regression (Different Degrees)')
plt.legend()
plt.show()

# Print MSE for each degree
for i, mse in enumerate(mse_values):
    print(f'Degree {i+1} MSE: {mse}')
```

```python
"""
Q14) Write a Python script that uses the Variance Inflation Factor (VIF) to check for multicollinearity in a dataset with multiple features.
Answer-
"""
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Load a sample dataset
from sklearn.datasets import make_regression
X, _ = make_regression(n_samples=100, n_features=5, noise=10, random_state=42)

# Convert the dataset into a DataFrame for easier analysis
df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])

# Add constant column for intercept
X_with_const = add_constant(df)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

print(vif_data)
```

```python
"""
Q16) Write a Python script that creates a machine learning pipeline with data standardization and a multiple linear regression model, and prints the R-squared score.
Answer-
"""
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression

# Generate synthetic data with multiple features
X, Y = make_regression(n_samples=100, n_features=5, noise=10, random_state=42)

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a pipeline with data standardization and linear regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', LinearRegression())
])

# Train the model
pipeline.fit(X_train, Y_train)

# Evaluate the model
score = pipeline.score(X_test, Y_test)
print(f'R-squared score: {score}')
```

```python
"""
Q21) Write a Python script that demonstrates how to serialize and deserialize machine learning models using joblib instead of pickling.
Answer-
"""
import joblib
from sklearn.linear_model import LinearRegression
import numpy as np

# Generate synthetic data
X = np.random.rand(100, 1) * 10
Y = 3 * X + 7 + np.random.randn(100, 1) * 5

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Serialize the model using joblib
joblib.dump(model, 'linear_model.pkl')

# Deserialize the model
loaded_model = joblib.load('linear_model.pkl')

# Print the model's coefficients and intercept
print(f'Coefficients: {loaded_model.coef_}')
print(f'Intercept: {loaded_model.intercept_}')
```

```python
"""
Q22) Write a Python script to perform linear regression with categorical features using one-hot encoding. Use the Seaborn 'tips' dataset.
Answer-
"""
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load the Seaborn 'tips' dataset
tips = sns.load_dataset('tips')

# Define features and target
X = tips[['sex', 'day', 'time']]
Y = tips['total_bill']

# Define a pipeline with OneHotEncoder for categorical features
pipeline = Pipeline([
    ('encoder', ColumnTransformer([('onehot', OneHotEncoder(), ['sex', 'day', 'time'])], remainder='passthrough')),
    ('regressor', LinearRegression())
])

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Fit the model
pipeline.fit(X_train, Y_train)

# Print the R-squared score
print(f'R-squared: {pipeline.score(X_test, Y_test)}')
```

```python
"""
Q23) Compare Ridge Regression with Linear Regression on a synthetic dataset and print the coefficients and R-squared score.
Answer-
"""
from sklearn.linear_model import Ridge, LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generate synthetic data with multiple features
X, Y = make_regression(n_samples=100, n_features=5, noise=10, random_state=42)

# Split the data into training and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Linear regression model
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)

# Ridge regression model
ridge_model = Ridge(alpha=1)
ridge_model.fit(X_train, Y_train)

# Print the coefficients and R-squared scores
print(f'Linear Regression Coefficients: {lr_model.coef_}')
print(f'R-squared (Linear Regression): {lr_model.score(X_test, Y_test)}')

print(f'Ridge Regression Coefficients: {ridge_model.coef_}')
print(f'R-squared (Ridge Regression): {ridge_model.score(X_test, Y_test)}')
```

```python
"""
Q24) Write a Python script that uses cross-validation to evaluate a Linear Regression model on a synthetic dataset.
Answer-
"""
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.datasets import make_regression

# Generate synthetic data
X, Y = make_regression(n_samples=100, n_features=5, noise=10, random_state=42)

# Create the linear regression model
model = LinearRegression()

# Perform 5-fold cross-validation and print the mean score
cv_scores = cross_val_score(model, X, Y, cv=5)
print(f'Cross-validation scores: {cv_scores}')
print(f'Mean cross-validation score: {cv_scores.mean()}')
```
